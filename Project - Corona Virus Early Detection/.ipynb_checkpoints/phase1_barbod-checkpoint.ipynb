```python
df_icu = df[df["ICU"]==1]
df_normal = df[df["ICU"]!=1]

print("OXYGEN_SATURATION_MEAN for icu and none icu: " , df_icu["OXYGEN_SATURATION_MEAN"].mean(),df_normal["OXYGEN_SATURATION_MEAN"].mean())
print("OXYGEN_SATURATION_MEDIAN for icu and none icu: " , df_icu["OXYGEN_SATURATION_MEDIAN"].mean(),df_normal["OXYGEN_SATURATION_MEDIAN"].mean())
print("OXYGEN_SATURATION_MIN for icu and none icu: " , df_icu["OXYGEN_SATURATION_MIN"].mean(),df_normal["OXYGEN_SATURATION_MIN"].mean())
print("OXYGEN_SATURATION_MAX for icu and none icu: " , df_icu["OXYGEN_SATURATION_MAX"].mean(),df_normal["OXYGEN_SATURATION_MAX"].mean())
print("OXYGEN_SATURATION_DIFF for icu and none icu: " , df_icu["OXYGEN_SATURATION_DIFF"].mean(),df_normal["OXYGEN_SATURATION_DIFF"].mean())
print("OXYGEN_SATURATION_DIFF_REL for icu and none icu: " , df_icu["OXYGEN_SATURATION_DIFF_REL"].mean(),df_normal["OXYGEN_SATURATION_DIFF_REL"].mean())
```

<div style="direction:rtl;line-height:300%;text-align:justify;" align="justify"><font face="B nazanin" size=4>
    <b style="font-size: 35px"></b><br>
از نتایج بالا به نظر می‌آید فیچر
    OXYGEN_SATURATION_MIN

    هم قابل توجه است.
    در قسمت زیر نمودار نسبت بیماران بستری در icu را نسبت به بیماران عادی برای

    به صورت 
    barchart 
    رسم شده است.
    تعداد افراد بستری در icu هم بالای هر bar نوشته شده است.
    برای 
    فیجر min
     همان طور که مشاهده می‌شود هرچه میزان این فیچر زیاد می‌شود نسبت بیماران icu به نرمال کاهش پیدا می‌کند.
    پس ظاهرا مقدار فیچر
    OXYGEN SATURATION MIN
    باید
    ارتباط بسیار زیادی با بستری  در icu داشته باشد.
      برای فیچر mean 
    نسبت افراد بستری در icu به افراد عادی برای مقادیر متفاوت mean 
    تفاوت چندانی نمی‌کند و تنها در حالت ۰.۹۵
    این نسبت مقدار قابل توجهی دارد اما چون تنها افراد بسیار کمی در این حد از اکسیژن هستند باز هم اطلاعات خوبی به ما نمی‌دهد.
    برای median هم دقیقا همین شرایط برقرار است.
    پس عملا دو فیچر
    OXYGEN SATURATION MEAN
    و
    OXYGEN SATURATION MEDIAN
    به صورت تنها اطلاعاتی در اختیار ما قرار نمی‌دهند اما دو فیچر MIN
    و 
    MAX
    به مقدار زیادی با احتمال بستری شدن در icu
    مرتبط هستند.
    با نتیجه‌گیری از همین نمودار‌ها سعی می‌کنیم تعداد فیچر‌ها را کمی کمتر کنیم. در بخش مهندسی ویژگی‌ها دو ستون MEAN و MEDIAN را برای اکسیژن حذف می‌کنیم.
 </font></div>

```python
bins = np.arange(0.6,1.05,0.05).tolist()
MIN_HEIGHT = 0.9

for column in ["OXYGEN_SATURATION_MEAN","OXYGEN_SATURATION_MEDIAN",
               "OXYGEN_SATURATION_MIN","OXYGEN_SATURATION_MAX"]:
    icu_numbers,_,_ = plt.hist(df_icu[column],bins=bins);
    icu_numbers[icu_numbers==0] = MIN_HEIGHT
    normal_numbers,_,_ = plt.hist(df_normal[column],bins=bins);
    normal_numbers[normal_numbers==0] = MIN_HEIGHT
    plt.close()
    proportion = icu_numbers/normal_numbers
    fig,ax = plt.subplots()
    ax.bar(bins[:-1],proportion,width=0.05,edgecolor='white');
    ax.set_xlabel(column);
    ax.set_ylabel("ICU patients / normal patients");
    rects = ax.patches
    
    
    for rect,icu_numbers in zip(rects,icu_numbers):
        height = rect.get_height()
        ax.text(rect.get_x() + rect.get_width() / 2, height+0.01 , icu_numbers,
                ha='center', va='bottom')

    plt.show()
```

```python
df["tags"].dropna()
all_diseases = ["Motor Neurone Disease","Smoker","Lung cancer","asthma","Kidney disease","heart disease"]
for c in all_diseases:
    df[c] = 0

for i in range(len(df)):
    s = df["tags"].iloc[i]
    if type(s)==str:
        diseases = s.split(", ")
        for d in diseases:
            df.loc[i,d]= 1

df_icu = df[df["ICU"]==1]
df_normal = df[df["ICU"]!=1]
df_normal
for d in all_diseases:
    fig,ax = plt.subplots()
    ax.bar([d+"=0", d+"=1"],[sum(df_icu[d]==0)/sum(df_normal[d]==0),sum(df_icu[d]==1)/sum(df_normal[d]==1)])
    ax.set_ylabel("icu / normal")
```

```python
clean_df = df.copy()
r = df.columns[8:-15].sort_values()
for i in range(0,len(r),5):
    x = df[r[i]].dropna().to_numpy()
    if np.var(x)==0:
        print(r[i] ," var 0")
        del clean_df[r[i]]
    for j in range(1,4):
        if sum((df[r[i+1]]-df[r[i+j+1]]).dropna()) == 0:
            print("columns " , r[i+1] ," and " , r[i+j+1] , " are the same")
            del clean_df[r[i+j+1]]

for f in ["MEAN","MEDIAN"]:
    del clean_df['BLOODPRESSURE_SISTOLIC_'+f]
    del clean_df['BLOODPRESSURE_DIASTOLIC_'+f]
    del clean_df['OXYGEN_SATURATION_'+f]
del clean_df['Smoker']
    
print("columns before: ", df.columns.shape[0])
print("columns after: ", clean_df.columns.shape[0])
```

    ستون AGE_PERCENTIL هم اطلاعات جامع‌تری را نسبت به AGE_ABOVE65 در اختیار قرار می‌دهد.
     مثلا دانستن این‌که فردی زیر ۲۰ سال است می‌تواند به ما خیلی کمک کند 
    که پیش‌بینی کنیم آن فرد در نهایت به ICU احتیاج پیدا می‌کند یا خیر.
    در ضمن داده‌های موجود در ستون
    AGE_PERCENTIL
    هم باید به صورت onehot
    دربیایند
    ،
    پس به ازای هر کدام از مقادیر یکتای آن یک ستون در نظر می‌گیریم که می‌تواند ۰ یا ۱ باشند.
    و در نهایت ستون AGE_PERCENTIL را پاک می‌کنیم.
    در حال حاضر تمامی ستون‌های دیتاست مقدار عددی دارند، ( به غیر از ستون window که هنگام دادن به مدل حذف می‌شود پس نیازی به عددی کردن آن نیست)
    
</font></div>

```python
del clean_df["AGE_ABOVE65"]
del clean_df["tags"]

for unique_value in pd.unique(clean_df["AGE_PERCENTIL"]):
    clean_df[unique_value] = 0
for unique_value in pd.unique(clean_df["AGE_PERCENTIL"]):
    clean_df[unique_value] = (clean_df["AGE_PERCENTIL"] == unique_value).astype(int)
del clean_df["AGE_PERCENTIL"]
del clean_df["WINDOW"]

```

<div style="direction:rtl;line-height:300%;text-align:justify;" align="justify"><font face="B nazanin" size=4>
    <b style="font-size: 35px"></b><br>
 حال نوبت به پر‌کردن خانه‌های
    خالی می‌رسد. 
    یکی از راه‌کار‌ها برای پر کردن خانه‌ی خالی 
    مربوط به فیچر f و بیمار x
    گذاشتن میانگین فیچر f
    برای بیمار x
    است.
    یعنی مثلا اگر ستون مربوط به اکسیژن یک بیمار در دو سطر حاوی عدد بود و در سه سطر مقدار nan را داشت
   ،
    میانگین اکسیژن این دو سطر را حساب می‌کنیم و در سه سطر دیگر قرار می‌دهیم.
 
</font></div>

```python
r = clean_df.groupby(by=["PATIENT_VISIT_IDENTIFIER"]).mean()
mean_column_values = pd.DataFrame(np.repeat(r.values,5,axis=0))
mean_column_values.columns = r.columns

for c in clean_df.columns:
    if c== "PATIENT_VISIT_IDENTIFIER" or c=="WINDOW":
        continue;
    clean_df.loc[clean_df[c].isna(),c]  = mean_column_values.loc[clean_df[c].isna(),c]

```

<div style="direction:rtl;line-height:300%;text-align:justify;" align="justify"><font face="B nazanin" size=4>
    <b style="font-size: 35px">تست مدل‌های مختلف:</b><br>
    
</font></div>

<div style="direction:rtl;line-height:300%;text-align:justify;" align="justify"><font face="B nazanin" size=4>
    <b style="font-size: 35px"></b><br>
۱۰ درصد دیتاست را برای تست و ۹۰ درصد باقی را برای ترینینگ در نظر می‌گیریم.
    چون سایز دیتاست کم است و همچنین چون ساختار مدل را بر اساس نتایج بدست‌آمده تغییر ندادم از قسمت 
    validation
    صرف نظر می‌کنیم.
    در این بخش از سه مدل معروف 
    decision tree
    ،
    random forest
    و
    logisitc regression
    که مناسب چنین دیتاستی هستند استفاده می‌کنیم و با توجه به نتایج یکی را برمی‌گزینیم.
    (دیتاست نسبتا کوچک است و مقادیر خالی نسبتا زیادی دارد. پس مدل‌های دیپ خیلی مناسب این مسئله نیستند چون احتمالا به خاطر تعداد بالای پارامتر‌هایشان overfit می‌شوند)
    یکی از اولین مدل‌هایی که برای دسته‌بندی داده‌ها با چنین حجمی به ذهن می‌آید decision tree می‌باشد.
    مدل اولی که تست کردیم مدل decision tree است که به 
    دقت ۰.۸۴، یادآوری ۰.۸۸ و f1_score 0.88 می‌رسد.
    (در حالتی که تنها با استفاده از سطر اول داده‌های بیمار، خروجی او را پیش‌بینی کند)
    اگر به درخت این مدل نگاهی بیندازید متوجه می‌شوید که اولین فیچر مورد توجه RESPIRATORY_RATE
    است که در بخش اول هم به علت اهمیت بالای آن به آن اشاره شد.
    در واقع اینکه RESPIRATORY_RATE_MEAN
    یکی از فیچر‌های اصلی برای خروجی باشد در ابتدا پیش‌بینی شده بود.
    فیچر اوره هم اهمیت قابل توجهی دارد.
</font></div>

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import f1_score
from sklearn.metrics import precision_score
from sklearn.metrics import recall_score
from sklearn.tree import plot_tree
from matplotlib import pyplot as plt

train_frac = 0.9
labels = clean_df.groupby(by=["PATIENT_VISIT_IDENTIFIER"]).max()
clean_df = clean_df.dropna()
main_labels = (labels.dropna())["ICU"].to_numpy()

final_df = clean_df.drop('Unnamed: 0',axis=1)
sample_train = 4
labels = np.repeat(main_labels,sample_train)
training_size = int(len(main_labels)*train_frac)

all_data = final_df.groupby(by="PATIENT_VISIT_IDENTIFIER").head(sample_train).drop(["PATIENT_VISIT_IDENTIFIER","ICU"],axis=1)
train_data = all_data[:training_size*sample_train]
train_label = labels[:training_size*sample_train]
clf = DecisionTreeClassifier(max_depth=15)
clf.fit(train_data, train_label)
print("Decision Tree:")
for i in range(1,2):
    all_data = final_df.groupby(by="PATIENT_VISIT_IDENTIFIER").head(i).drop(["PATIENT_VISIT_IDENTIFIER","ICU"],axis=1)
    test_data= all_data[training_size*i:]
    test_label = np.repeat(main_labels[training_size:],i)
    
    model_output =  clf.predict(test_data)
    f1score = f1_score(test_label,model_output)
    precision = precision_score(test_label,model_output)
    recall = recall_score(test_label,model_output)
    print(i , "samples")
    print("f1 score: ",f1score)
    print("precision: ", precision)
    print("recall: " , recall)
    print("_________________________________________________________")
    
plt.rcParams["figure.figsize"] = (100,100)
fig,ax = plt.subplots()
ax = plot_tree(clf,max_depth = 5,feature_names=all_data.columns)
```

<div style="direction:rtl;line-height:300%;text-align:justify;" align="justify"><font face="B nazanin" size=4>
    <b style="font-size: 35px"></b><br>
مدل دیگری که می‌توانیم استفاده کنیم حالت پیچیده‌تر decision tree یا همان random forest
است.
    همان‌طور که از نتایج پیداست این مدل عملکرد بهتری نسبت به decision tree دارد.
    در کل random forest نسبت به decision tree پیچیده‌تر است و generalization بیشتری دارد.
    شاید تنها برتری decision tree سرعت آن باشد که در این مساله خیلی مورد توجه نیست چون سایز دیتاست کوچک است و fit شدن زمان زیادی نمی‌برد.
    به همین علت و با توجه به نتایج بهتر random forest فعلا این مدل را به عنوان برترین مدل انتخاب می‌کنیم.
</font></div>

```python
from sklearn.ensemble import RandomForestClassifier

random_forest = RandomForestClassifier(max_depth=15)
random_forest.fit(train_data, train_label)
print("Random Forest: ")

for i in range(1,2):
    all_data = final_df.groupby(by="PATIENT_VISIT_IDENTIFIER").head(i).drop(["PATIENT_VISIT_IDENTIFIER","ICU"],axis=1)
    test_data= all_data[training_size*i:]
    test_label = np.repeat(main_labels[training_size:],i)
    
    model_output =  random_forest.predict(test_data)
    f1score = f1_score(test_label,model_output)
    precision = precision_score(test_label,model_output)
    recall = recall_score(test_label,model_output)
    print(i , "samples")
    print("f1 score: ",f1score)
    print("precision: ", precision)
    print("recall: " , recall)
    print("_________________________________________________________")
```

<div style="direction:rtl;line-height:300%;text-align:justify;" align="justify"><font face="B nazanin" size=4>
    <b style="font-size: 35px"></b><br>
مدل
   logistic regression
    از دو مدل قبلی عملکرد بهتری دارد.
    این بهبود عملکرد به علت یادآوری بالاتر است.
    در این مساله چون رابطه‌ی بین فیچر‌ها و لیبل نهایی خیلی پیچیده نیست نیازی به استفاده از مدل‌های پیچیده وجود ندارد و یک مدل خطی ساده به راحتی با دقت بالا می‌تواند این کار را انجام دهد.
    این مدل از random forest
    پیچیدگی کمتری دارد و برای چنین دیتاستی مناسب‌تر است به همین علت نتایج بهتری هم کسب کرده است.

```python
from sklearn.linear_model import LogisticRegression
print("Logistic Regression:")
logistic_model = LogisticRegression(max_iter=10000).fit(train_data, train_label)
for i in range(1,2):
    all_data = final_df.groupby(by="PATIENT_VISIT_IDENTIFIER").head(i).drop(["PATIENT_VISIT_IDENTIFIER","ICU"],axis=1)
    test_data= all_data[training_size*i:]
    test_label = np.repeat(main_labels[training_size:],i)
    model_output =  logistic_model.predict(test_data)
    f1score = f1_score(test_label,model_output)
    precision = precision_score(test_label,model_output)
    recall = recall_score(test_label,model_output)
    print(i , "samples")
    print("f1 score: ",f1score)
    print("precision: ", precision)
    print("recall: " , recall)
    print("_________________________________________________________")
```
