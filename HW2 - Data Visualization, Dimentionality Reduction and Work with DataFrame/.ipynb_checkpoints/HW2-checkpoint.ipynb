```python
pca_dataset=dataset[['Age','Sex','Housing','Saving accounts','Checking account','Credit amount','Duration','Purpose']]
pca_dataset.head(20)
```

<div dir=rtl>
    ۱۷.(۶ نمره)
    یکی از مواردی که ممکن است تاثیر مخربی روی تحلیل ما بگذارد تفاوت اندازه‌های اعداد ستون‌های مختلف است(چرا؟ خیلی کوتاه توضیح دهید) برای همین بهتر است داده‌ها را نرمالایز کنیم.
</div>

```python
for column in pca_dataset.columns: 
   pca_dataset[column] = (pca_dataset[column] - pca_dataset[column].mean()) / pca_dataset[column].std()     
  
pca_dataset.head(20)
```

**دلایل اهمیت نرمال کردن داده‌ها**

1) افزایش سرعت آنالیز داده‌ها و نیز سرعت سورت کردن داده‌‌ها

2)راحتی انجام عملیات بر روی داده‌ها و آنالیز کردن و درک آن‌ها

3) صرفه‌جو‌یی در حافظه مورد نیاز برای نگه‌داری داده‌ها

4)عدم از دست رفتن دقت داده‌ها به علت تفاوت زیاد؛ برای مثال اگر داده‌های خیلی 
بزرگ یا خیلی کوچک داشته باشیم ، ممکن است دقت داده‌های خیلی کوچک را از دست بدهیم

5) از بین بردن داده‌‌های تکراری و حذف داده‌های ناقص و یا یکسان‌سازی تمام داده‌ها
(برای مثال ممکن است داده‌ها از دیتابیس‌های متفواتی به‌دست آمده باشند و یکدست نباشند)

6)حذف کردن موارد ناقص در داده‌ها و یا پر کردن آنها

مقدار و بردار ویژه‌ی ماتریس کواریانس را بدست بیاورید.

```python
import scipy.linalg as la

eigvals, eigvec = la.eig(cov_matrix)
print(eigvals)
print()
print(eigvec)
```

<div dir=rtl>
    ۲۱.(۳ نمره)
    برای انتخاب اجزای اساسی(Principle Components) سه کار ریز را انجام دهید:
    <br\>
هر بردار ویژه را با مقدار ویژه متناظرش مربوط کنید (می‌توانید بصورت جفتیpair آن‌ها را کنار هم قرار دهید)
        <br\>
            لیستی که بردار و مقدار ویژه‌ها دارید را بر اساس مقدار ویژه‌ها مرتب کنید
            <br\>
                مقدار ویژه‌های مرتب شده را به ترتیب نزولی در خروجی نمایش دهید
</div>

```python
eigens = [(eigvals[i], eigvec[:,i]) for i in range(len(eigvals))]
print("eigenvector respected to each eigen value")
print(eigens)
print()
eigens.sort(reverse=True)
print("sorted values of eigen values")
print([eigens[i][0] for i in range(len(eigens))])
```

<div dir=rtl>
    ۲۲.(۴ نمره)
همه‌ی
    explained variance 
    را بدست بیاورید و در نمودار pie نمایش دهید
    حتما درصد هرکدام از اسلایس‌ها در نمودار مشخص شده‌باشد.
</div>

```python
sorted_eigvalues = [eigens[i][0] for i in range(len(eigens))]
print("sorted eigenvalues",sorted_eigvalues)

explained_variance= []
for i in range(len(sorted_eigvalues)):
     explained_variance.append((sorted_eigvalues[i]/sum(sorted_eigvalues)))        
print("explained variance",explained_variance)


sizes = explained_variance
fig1, ax1 = plt.subplots()
ax1.pie(sizes, autopct='%1.1f%%', shadow=True, startangle=90)
ax1.axis('equal')  
plt.show()
```

با توجه به آنچه در بخش قبل دیدید تعداد کامپوننت‌ها را طوری انتخاب کنید که پوشش حداقل ۶۰٪ای از اطلاعات داشته باشد و سپس projection matrix را با استفاده از آن‌ها تشکیل دهید.
    </div>

```python
var_comp_sum = np.cumsum(explained_variance)/sum(explained_variance)
print("Cumulative proportion of explained variance: \n%s" %var_comp_sum)
number_of_components=0
for i in range(len(var_comp_sum)):
  if(var_comp_sum[i]>=0.6):
    number_of_components=i+1
    break
print("number_of_components",number_of_components)

sorted_eigvectors = [eigens[i][1] for i in range(len(eigens))]
sorted_eigvactors_matrix=np.array(sorted_eigvectors)
projection_matrix = (sorted_eigvactors_matrix.T[:][:number_of_components]).T
print("projection_matrix\n",projection_matrix)
```

<div dir=rtl>
    ۲۴.(۵ نمره)
        با استفاده از ماتریس بدست آمده در مرحله قبل X را کاهش دهید
    و در یک نمودار scatter که x و y آن ازpricipal componentهای اول و دوم شما بدست آمده‌اند بر حسب Risk نتایج را نشان دهید 
</div>

```python
pca_dataset_matrix=pca_dataset.values
X_pca = pca_dataset_matrix.dot(projection_matrix)
print("Projection Matrix")
print(X_pca)
X_pca_T=X_pca.T
new_df_for_pca=pd.DataFrame() 
new_df_for_pca.insert(0,"first component",X_pca_T[0],True)
new_df_for_pca.insert(1,"second component",X_pca_T[1],True)
new_df_for_pca.insert(2,"Risk",dataset['Risk'],True)
sns.scatterplot(data=new_df_for_pca, x='first component', y='second component', hue='Risk');
```
