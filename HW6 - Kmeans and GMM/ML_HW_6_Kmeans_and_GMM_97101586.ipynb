# First Data set - 2 dimentioanl

```python
from sklearn.datasets.samples_generator import make_classification
X, Y = make_classification(n_samples=700, n_features=2,
                           n_informative=2, n_redundant=0,
                           n_classes=2)

#X is the 2 dimentional data
#Y is the labels 
```

```python
plt.scatter(X[:,0],X[:,1],c='red',label='unclustered data')
plt.xlabel('X1')
plt.ylabel('X2')
plt.legend()
plt.title('Plot of data points')
plt.show()
```

```python
class Kmeans:

    def __init__(self, k , data):
        self.k = k
        self.data = data
    
    def reassignment(self, centroids):
      cluster_of_each_point = []
      for d in data:
        cluster_of_each_point .append(np.argmin( np.sum( (d.reshape(1, data.shape[1])  -  centroids) **2 , axis =1 )))
      return cluster_of_each_point
    
    def update_centroids(self, cluster_of_each_point):
      centroids = []
      for cluster in range(k):
        centroids.append(np.mean( [data[d] for d in range(data.shape[0]) if cluster_of_each_point[d] == cluster] , axis = 0 ))
      return centroids

    def convergence(self , previous_centroids, new_centroids):
      previous_centroids = np.array(previous_centroids)
      previous_centroids.reshape(2,data.shape[1])
      new_centroids = np.array(new_centroids)
      new_centroids.reshape(2,data.shape[1])
      difference = new_centroids - previous_centroids
      flag = True
      for i in range(2):
        for j in range(data.shape[1]):
          if abs(difference[i][j]) >= 0.0000001:
            flag = False
      return flag
```

```python
k = 2 
data = X
data = np.array(data)
centroids = []
number1 = random.randrange(0, data.shape[0])
number2 = random.randrange(0, data.shape[0])
while number2 == number1 :
  number2 = random.randrange(0, data.shape[0])
centroid1 , centroid2 = X[number1] , X[number2]
centroids.append(centroid1)
centroids.append(centroid2)
centroids = np.array(centroids)

k_means = Kmeans(k , data)
step = 0 
while True:
  step +=1
  cluster_of_each_point = k_means.reassignment(centroids)
  new_centroids = k_means.update_centroids(cluster_of_each_point)
  if k_means.convergence( centroids , new_centroids):
    break
  else:
    centroids = new_centroids

centroids = np.array(centroids)
centroids.reshape(2,2)
print("cluster_of_each_point",cluster_of_each_point)
print("centroids", centroids)
print("steps", step)
```

```python
def get_confusion_matrix(y_true, y_pred):
    n_classes = len(np.unique(y_true))
    conf = np.zeros((n_classes, n_classes))
    for actual, pred in zip(y_true, y_pred):
        conf[int(actual)][int(pred)] += 1
    return conf.astype('int')
confusion_matrix = get_confusion_matrix(Y, cluster_of_each_point)
print(confusion_matrix)
```

```python
# Rand Index = ( TP + TN ) / ( TP + TN + FP + FN )
RI = (confusion_matrix[0][0] + confusion_matrix[1][1] ) / (confusion_matrix[0][0] + confusion_matrix[0][1] + confusion_matrix[1][0]  + confusion_matrix[1][1] ) 
print(RI)
```

```python
#for each cluster ci, select the maximum value from its row, sum them together and finally divide by the total number of data points
purity=np.sum(np.amax(confusion_matrix, axis=0)) / np.sum(confusion_matrix)
print(purity)
```

**GMM**

```python
class GaussianMixtureModel:
    def __init__(self, k, Mu):
        self.k = k
        self.Pi = None
        self.Mu = Mu
        self.Sigma = None

    def parameter_initialization(self, X, random_state=None):
        n_samples, n_features = np.shape(X)
        rnd = np.random.RandomState(seed=random_state)
        self.Pi = np.ones(self.k) / self.k
        self.Sigma = np.tile(np.diag(np.var(X, axis=0)), (self.k, 1, 1))

    def calculate_Nmat(self, X):
        '''
        Method for calculating array corresponding $\mathcal{N}(x_n | \mu_k)$
        Nmat : 2-D numpy array for probability density of each sample and component. Nmat[n, k] = $\mathcal{N}(x_n | \mu_k)$.
        '''
        n_samples, n_features = np.shape(X)
        Diff = np.reshape(X, (n_samples, 1, n_features)) - np.reshape(self.Mu, (1, self.k, n_features))
        L = np.linalg.inv(self.Sigma)
        exponent = np.einsum("nkj,nkj->nk", np.einsum("nki,kij->nkj", Diff, L), Diff)
        Nmat = np.exp(-0.5 * exponent) / np.sqrt(np.linalg.det(self.Sigma)) / (2 * np.pi) ** (n_features / 2)
        return Nmat

    def Estep(self, X):
        '''
        Method for calculating the array corresponding to responsibility.
        Nmat : 2-D numpy array for probability density of each sample and component. Nmat[n, k] = $\mathcal{N}(x_n | \mu_k)$.
        '''
        n_samples, n_features = np.shape(X)
        Nmat = self.calculate_Nmat(X)
        probabilities = Nmat * self.Pi
        Gama = probabilities / np.reshape(np.sum(probabilities, axis=1), (n_samples, 1))
        return Gama

    def Mstep(self, X, Gama):
        '''
        Method for calculating the model parameters based on the responsibility gamma.
        Gama : 2-D numpy array representing responsibility of each component for each sample in X,
        where Gamt[n, k] = $\gamma_{n, k}$.
        '''
        n_samples, n_features = np.shape(X)
        Diff = np.reshape(X, (n_samples, 1, n_features)) - np.reshape(self.Mu, (1, self.k, n_features))
        Nk = np.sum(Gama, axis=0)
        self.Pi = Nk / n_samples
        self.Mu = Gama.T @ X / np.reshape(Nk, (self.k, 1))
        self.Sigma = np.einsum("nki,nkj->kij", np.einsum("nk,nki->nki", Gama, Diff), Diff) / np.reshape(Nk,
                                                                                                       (self.k, 1, 1))

    def calculate_probability_density(self, X):
        '''
        Method for calculating the probablity density $\sum_k \pi_k \mathcal{N}(x_n | \mu_k)$
        '''
        prob_density = self.calculate_Nmat(X) @ self.Pi
        return prob_density


    def fit(self, X, treshold , random_state=0):
        self.parameter_initialization(X, random_state=random_state)
        log_likelihood = - np.float("inf")
        log_likelihood_old = 0
        steps = 0
        while abs(log_likelihood - log_likelihood_old) >= treshold:
            steps += 1
            Gama = self.Estep(X)
            self.Mstep(X, Gama)
            log_likelihood_old = log_likelihood
            log_likelihood = np.sum(np.log(self.calculate_probability_density(X)))
        print("steps", steps)
        print(f"log_likelihood : {log_likelihood}")


    def predict(self, X):
        prediction = np.argmax(self.Estep(X), axis=1)
        return prediction
```

```python
def get_meshgrid(x, y, nx, ny, margin=0.1):
    x_min, x_max = (1 + margin) * x.min() - margin * x.max(), (1 + margin) * x.max() - margin * x.min()
    y_min, y_max = (1 + margin) * y.min() - margin * y.max(), (1 + margin) * y.max() - margin * y.min()
    xx, yy = np.meshgrid(np.linspace(x_min, x_max, nx), np.linspace(y_min, y_max, ny))
    return xx, yy
xx, yy = get_meshgrid(X[:, 0], X[:, 1], nx=100, ny=100, margin=0.1)
```

```python
def plot_predicted_label(ax, clf, xx, yy, X, t):
    Z = clf.predict(np.c_[xx.ravel(), yy.ravel()])
    Z = Z.reshape(xx.shape)
    ax.contourf(xx, yy, Z, alpha=0.7)
    ax.scatter(X[:,0], X[:,1], c=t, edgecolor='k')
    
def plot_prob_density(ax, model, xx, yy, X, t):
    Z = model.calculate_probability_density(np.c_[xx.ravel(), yy.ravel()])
    Z = Z.reshape(xx.shape)
    ax.scatter(X[:,0], X[:,1], c=t, edgecolor='k')
    ax.contour(xx, yy, Z)
    
def plot_result(model, xx, yy, X, t):
    fig = plt.figure(figsize=(8,4))
    ax = fig.add_subplot(121)
    plot_predicted_label(ax, model, xx, yy, X, t)
    ax = fig.add_subplot(122)
    plot_prob_density(ax, model, xx, yy, X, t)
```
